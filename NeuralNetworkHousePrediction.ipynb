# Importing Libraries





```
import keras
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Importing Dataset and Checkout

```
#If you want to import from Drive
df=pd.read_csv('/content/USA_Housing.csv')
df.head()
```

# Checking Insights of data

```
df.info()
```

# Get Data Ready For Training a Linear Regression Model

```
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]

y = df[['Price']]
```

# Split Data into Train, Test

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

# Creating and Training the Neural Network Model

```
model = keras.models.Sequential()

model.add(keras.layers.Dense(5, activation='relu', input_shape=(5,)))
model.add(keras.layers.Dense(5, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

# Predictions from our Neural Network Model

```
import numpy as np
model.fit(X_train, y_train , epochs=30, callbacks=[keras.callbacks.EarlyStopping(patience=3)])
```

```
predictions =model.predict(X_test)
```

```
plt.scatter(y_test,model.predict(X_test), color='red')
```

**Finally, if you want to save this model and use it for later, use the following code. You can see it saved to our colab environment**

```
sns.distplot((y_test-model.predict(X_test)),bins=50);
```

# Regression Evaluation Metrics

```
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print("R2 score =", round(metrics.r2_score(y_test, predictions), 2))
```
